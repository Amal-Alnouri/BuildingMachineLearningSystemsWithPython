# The Iris dataset

## visualization

```python
from matplotlib import pyplot as plt
from sklearn.datasets import load_iris
import numpy as np
```

### Load data from scikit learn

```python
# We load the data with load_iris from sklearn
data = load_iris()
features = data['data']
feature_names = data['feature_names']
target = data['target']
target_names = data['target_names']
```

### Plot data

```python
def plot_data(f1, f2, subplt):
    for t,marker,c in zip(range(3),">ox","rgb"):
        # We plot each class on its own to get different colored markers
        plt.subplot(subplt)
        plt.scatter(features[target == t,f1], features[target == t, f2], marker=marker, c=c, s=10)
        plt.xlabel(feature_names[f1])
        plt.ylabel(feature_names[f2])
        
plot_data(0, 1, 231) # 2 rows 3 columns 1 subplot number (max_number=rows*cols)
plot_data(0, 2, 232)
plot_data(0, 3, 233)
plot_data(1, 2, 234)
plot_data(1, 3, 235)
plot_data(2, 3, 236)
plt.tight_layout() # adjust spacing between subplots to minimize the overlaps.
plt.show()

# Refer to https://matplotlib.org/users/tight_layout_guide.html for more information about subplots
```

## Building Classification Model

#### The petal length seems to be able to separate Iris Setosa from the other two flower species on its own. This bit of code discovers where the cutoff is.

```python
plength = features[:, 2]
# use numpy operations to get setosa features
is_setosa = target == 0
# This is the important step:
max_setosa =plength[is_setosa].max()
min_non_setosa = plength[~is_setosa].min()

print('--------------------------------\nModel\n--------------------------------')
print('Maximum of setosa: {0}.'.format(max_setosa))
print('Minimum of others: {0}.'.format(min_non_setosa))

def simlpe_setosa_detection(features):
    if features[2] < 2: print('Iris Setosa')
    else: print('Iris Virginica or Iris Versicolour')

print('--------------------------------\ntest\n--------------------------------')
import random
sample = random.randint(0, len(features))
real_label = target_names[target[sample]]
print('test sample: {}'.format(features[sample]))
print('correct label: {}'.format(real_label))
print('prediction:')
simlpe_setosa_detection(features[sample])
```
